Import libary yang digunakan

```python
import numpy as np
import pandas as pd
```

Membuat data frame yang digunakan untuk train

```python
data_train = pd.read_csv('train.csv')
data_train
```

```python
data_train.info()
```

Mengolah data frame yang sudah ada, membuang data-data yang tidak berpengaruh terhadap survival penumpang dan data yang kosong. Memisahkan kolom 'Survived' dari kolom yang lain.

```python
data_train.drop('PassengerId', axis = 1, inplace = True)
data_train.drop('Name', axis = 1, inplace = True)
data_train.drop('Cabin', axis = 1, inplace = True)
data_train.drop('Ticket', axis = 1, inplace = True)
data_train.drop('Embarked', axis = 1, inplace = True)
data_train.dropna(subset = ['Age'], inplace = True)
data_train['Fare'] = np.round(data_train['Fare'], 2)
survived_train = data_train['Survived']
data_train.drop('Survived', axis = 1, inplace = True)
data_train.head()
```

```python
survived_train.head()
```

```python
data_train.info()
```

Mengubah kolom 'Sex' menjadi kolom biner.

```python
sex_binary= pd.get_dummies(data_train['Sex'])
data_train['female'] = sex_binary['female']
data_train['male'] = sex_binary['male']
data_train.drop('Sex', axis = 1, inplace = True)
```

```python
data_train
```

Menginputkan nilai X dan y.

```python
X = data_train[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'female', 'male']]
y = survived_train
```

Melakukan train terhadap model yang dibuat.

```python
from sklearn.tree import DecisionTreeClassifier
train_model = DecisionTreeClassifier()

train_model.fit(X,y)
```

Menguji model yang dibuat dengan data yang ada pada data frame untuk train.

```python
sample_data = X.iloc[0,:].tolist()
print(sample_data)
```

```python
train_model.predict([sample_data])
```

Menunjukkan akurasi model yang dibuat.

```python
train_model.score(X,y)
```

Membuat data frame yang digunakan untuk mengetest model yang dibuat.

```python
data_test = pd.read_csv('test.csv')
data_test
```

Mengolah data yang digunakan untuk test, membuat data frame dengan kolom yang sama seperti yang digunakan untuk menguji. Membuang data-data yang bernilai kosong.

```python
data_test.drop('PassengerId', axis = 1, inplace = True)
data_test.drop('Name', axis = 1, inplace = True)
data_test.drop('Cabin', axis = 1, inplace = True)
data_test.drop('Ticket', axis = 1, inplace = True)
data_test.drop('Embarked', axis = 1, inplace = True)
data_test.dropna(subset = ['Age'], inplace = True)
data_test['Fare'] = np.round(data_test['Fare'], 2)
sex_binary_test= pd.get_dummies(data_test['Sex'])
data_test['female'] = sex_binary_test['female']
data_test['male'] = sex_binary_test['male']
data_test.drop('Sex', axis = 1, inplace = True)
data_test.dropna(subset = ['Fare'], inplace = True)
data_test.head()
```

```python
data_test.info()
```

Menerapkan model prediksi yang telah dibuat sebelumnya ke data 'test'.

```python
pred_results = train_model.predict(data_test)
```

Memasukkan data 'Survived' ke copy data frame test untuk digunakan saat menguji single data point.

```python
data_results = data_test.copy()
data_results['Survived'] = pred_results
data_results.head()
```

Melakukan prediksi single data point.

```python
single_data_point = data_test.iloc[2,:].tolist()
train_model.predict([single_data_point])
```
